# Making 3D Brain Views without Opening a Browser

Pycortex can render 3D brain views without opening a browser using
`cortex.export.plot_panels` with `headless=True`. This uses
Playwright to launch a headless Chromium browser that renders
the WebGL scene.

## Prerequisites

Install Playwright and download the bundled Chromium binary:

```bash
pip install playwright
playwright install chromium
```

## Composite view with all panels

The `params_flatmap_inflated_lateral_medial_ventral` preset
renders a flatmap with lateral, medial, and ventral insets —
all in a single figure.

````python
import cortex
import cortex.export
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
volume = cortex.Volume.random(subject="S1", xfmname="fullhead")

fig = cortex.export.plot_panels(
    volume,
    headless=True,
    **cortex.export.params_flatmap_inflated_lateral_medial_ventral,
)
plt.show()
````

## Other predefined layouts

Pycortex ships with several ready-made panel configurations.
Here are all the available presets:

````python
predefined = {
    name: getattr(cortex.export, name)
    for name in sorted(dir(cortex.export))
    if name.startswith("params_")
}

for name, params in predefined.items():
    fig = cortex.export.plot_panels(volume, headless=True, **params)
    w, h = fig.get_size_inches()
    fig.suptitle(name, fontsize=14, fontweight="bold", y=1.02)
    fig.subplots_adjust(top=0.95)
    plt.show()
````
